# Train Test Split

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df, test_size=0.1, random_state=42)
```

# Column Wise preliminary Analysis

```python
# Lowercasing column names
df_train.columns = df_train.columns.str.lower()
# Renaming columns to remove spaces and special characters
df_train.columns = df_train.columns.str.replace(' ', '_').str.replace('-', '_').str.replace('.', '_')
# Lowercasing column names in the test set
df_test.columns = df_test.columns.str.lower()
# Renaming columns in the test set to match the train set
df_test.columns = df_test.columns.str.replace(' ', '_').str.replace('-', '_').str.replace('.', '_')
# Displaying the first few rows of the train set
df_train.head()
```

## AVG_Bill_amt

```python
df_train['avg_bill_amt'].describe()
```

```python
# make a new column named 'avg_bill_amt' which is the average of bill_amt1 to bill_amt6
df_train['avg_bill_amt_check'] = df_train[['bill_amt1', 'bill_amt2', 'bill_amt3', 'bill_amt4', 'bill_amt5', 'bill_amt6']].mean(axis=1)
df_train[['avg_bill_amt', 'avg_bill_amt_check']]
```

```python
# print those columns where difference between avg_bill_amt and avg_bill_amt_check is greater than 1

df_train[abs(df_train['avg_bill_amt'] - df_train['avg_bill_amt_check']) > 1][['avg_bill_amt', 'avg_bill_amt_check']]
```

due to this disparity between the columns we will keep the avg_bill_amt_check as our avg and remove the other

```python
# remove avg_bill_amt and rename avg_bill_amt_check to avg_bill_amt
df_train.drop(columns=['avg_bill_amt'], inplace=True)
df_train.rename(columns={'avg_bill_amt_check': 'avg_bill_amt'}, inplace=True)
```

```python
# round avg_bill_amt to 2 decimal places
df_train['avg_bill_amt'] = df_train['avg_bill_amt'].round(2)
# check if avg_bill_amt has any negative values
df_train[df_train['avg_bill_amt'] < 0]
```

```python
# calculate avg_bill_amt for the test set
df_test['avg_bill_amt_check'] = df_test[['bill_amt1', 'bill_amt2', 'bill_amt3', 'bill_amt4', 'bill_amt5', 'bill_amt6']].mean(axis=1)
# remove avg_bill_amt and rename avg_bill_amt_check to avg_bill_amt
df_test.drop(columns=['avg_bill_amt'], inplace=True)
df_test.rename(columns={'avg_bill_amt_check': 'avg_bill_amt'}, inplace=True)
# round avg_bill_amt to 2 decimal places
df_test['avg_bill_amt'] = df_test['avg_bill_amt'].round(2)
```

## PAY to Bill ratio

```python
df_train['pay_to_bill_ratio'].describe()
```

```python
# creating two derived columns called total_payment and total_bill over 6 months
df_train['total_payment'] = df_train[['pay_amt1', 'pay_amt2', 'pay_amt3', 'pay_amt4', 'pay_amt5', 'pay_amt6']].sum(axis=1)
df_train['total_bill'] = df_train[['bill_amt1', 'bill_amt2', 'bill_amt3', 'bill_amt4', 'bill_amt5', 'bill_amt6']].sum(axis=1)
# calculate pay_to_bill_ratio   
df_train['pay_to_bill_ratio'] = df_train['total_payment'] / (df_train['total_bill']+0.0001)  # adding a small value to avoid division by zero
# round pay_to_bill_ratio to 2 decimal places
df_train['pay_to_bill_ratio'] = df_train['pay_to_bill_ratio'].round(2)
# check if pay_to_bill_ratio has any negative values
df_train[df_train['pay_to_bill_ratio'] < 0]
```

```python
df_train['pay_to_bill_ratio'].describe()
```

```python
# do the same for the test set
df_test['total_payment'] = df_test[['pay_amt1', 'pay_amt2', 'pay_amt3', 'pay_amt4', 'pay_amt5', 'pay_amt6']].sum(axis=1)
df_test['total_bill'] = df_test[['bill_amt1', 'bill_amt2', 'bill_amt3', 'bill_amt4', 'bill_amt5', 'bill_amt6']].sum(axis=1)
# calculate pay_to_bill_ratio
df_test['pay_to_bill_ratio'] = df_test['total_payment'] / (df_test['total_bill'] + 0.0001)  # adding a small value to avoid division by zero
# round pay_to_bill_ratio to 2 decimal places
df_test['pay_to_bill_ratio'] = df_test['pay_to_bill_ratio'].round(2)
# check if pay_to_bill_ratio has any negative values
df_test[df_test['pay_to_bill_ratio'] < 0]
```
